```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

# Assuming your DataFrame is named df

# Selecting the numerical features for comparison
style_features = ['popularity', 'duration_ms', 'danceability', 'energy', 'key', 
                      'loudness', 'mode', 'speechiness', 'acousticness', 
                      'instrumentalness', 'liveness', 'valence', 'tempo', 
                      'time_signature']

numerical_features = ['danceability', 'energy', 'valence', 'tempo', 
                  'acousticness', 'popularity', 'loudness', 
                  'speechiness']

# Scaling the features to normalize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[numerical_features])

# Initialize the NearestNeighbors model
nbrs = NearestNeighbors(n_neighbors=9, algorithm='auto').fit(scaled_features)

# Find the nearest neighbors for the first song (index 0)
distances, indices = nbrs.kneighbors([scaled_features[0]])

# Get the indices of the top 8 closest songs, excluding the first song itself
closest_indices = indices[0][1:9]

# Display the closest tracks
closest_tracks = df.iloc[closest_indices][['artists', 'track_name'] + numerical_features]
print(closest_tracks)
```
